
#### Write a function in python to count the number 0f lines from a text file "story.txt" ,which is not starting with an alphabet "T"
 
Eg :- Story.txt file contains the following lines.
    
    A boy is playing there.
    There is a playground.
    An aeroplane is in the sky.
    The sky is pink.


```python
file = open("story.txt","r")
count=0
for word in file:
    if word [0]!='T':    # ! use for not Startinh
        count=count+1
print("Number of Line which is Not Starting With --->T: ",count)
file.close()
```

#### 4) Write a function in python to read lines from a text file "notes.txt".your function should find and display the occurrence of the  word "the".
Eg :- file content 
    
    "India,is the fastest-growing economy.India is looking for more investment around the globe.The whole world is looking at India as a great market.Most of the Indians can foresee the hright that India is capable of reachig"

```python
file = open("notes.txt","r")
count = 0
data = file.read()
words = data.split()
for word in words:
    if word =="the" or word =="The":
        count += 1
print("The occurrence of the word ,the-->",count)
file.close()
```
