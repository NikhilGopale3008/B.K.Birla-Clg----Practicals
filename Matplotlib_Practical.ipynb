#### 3.	Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

#### 6.	Calculate total sales data for last year for each product and show it using a Pie chart

```python
month = com['month_number']

label=['facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer']

data = [com['facecream'].sum(),com['facewash'].sum(),com['toothpaste'].sum(),
        com['bathingsoap'].sum(),com['shampoo'].sum(), com['moisturizer'].sum()]

plt.pie(data,labels=label,autopct='%0.1f%%')
plt.axis("equal")
plt.show()
```

#### 7.	Read Bathing soap facewash of all months and display it using the Subplot

```python
monthList  = com['month_number'].tolist()
bathingsoap   = com['bathingsoap'].tolist()
faceWashSalesData   = com['facewash'].tolist()
f, axarr = plt.subplots(2, sharex=True,figsize=(12,7))
axarr[0].plot(monthList, bathingsoap, label = 'Bathingsoap Sales Data', color='blue', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(monthList, faceWashSalesData, label = 'Face Wash Sales Data', color='green', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xticks(monthList)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

#### 8. Read all product sales data and show it using the stack plot

```python
monthList  = com['month_number'].tolist()

faceCremSalesData   = com['facecream'].tolist()
faceWashSalesData   = com['facewash'].tolist()
toothPasteSalesData = com['toothpaste'].tolist()
bathingsoapSalesData   = com['bathingsoap'].tolist()
shampooSalesData   = com['shampoo'].tolist()
moisturizerSalesData = com['moisturizer'].tolist()
plt.figure(figsize=(12,7))
plt.plot([],[],color='red', label='face Cream', linewidth=5)
plt.plot([],[],color='blue', label='Face wash', linewidth=5)
plt.plot([],[],color='green', label='Tooth paste', linewidth=5)
plt.plot([],[],color='black', label='Bathing soap', linewidth=5)
plt.plot([],[],color='yellow', label='Shampoo', linewidth=5)
plt.plot([],[],color='grey', label='Moisturizer', linewidth=5)

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['red','blue','green','black','yellow','grey'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in')
plt.title(' product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
